```python
import pandas as pd
file_dir = "/content/gdrive/MyDrive/AI개발팀/jhjin/data/"
file_name = "drugDB3.xlsx"
file_path = os.path.join(file_dir, file_name)
df = pd.read_excel(file_path)
df.head()
```

```python
from llama_index.core import KeywordTableIndex, SimpleDirectoryReader, VectorStoreIndex, ServiceContext
from llama_index.llms.openai import OpenAI
```

```python
# alternatively
# from langchain.llms import ...

documents = SimpleDirectoryReader(file_dir).load_data()
```

```python
documents
```

```python
# 빈 엑셀파일 1개만 포함된 documents로 index 생성 시 1초 소요, $0.02 소모.
index1 = VectorStoreIndex.from_documents(documents)
```

```python
llm = OpenAI(model="gpt-4-1106-preview", api_key=OPENAI_API_KEY)
service_context = ServiceContext.from_defaults(llm=llm)
index1 = VectorStoreIndex.from_documents(documents, service_context=service_context, show_progress=True)
```
